```python
import nltk
from nltk import word_tokenize, pos_tag
from nltk.chunk import ne_chunk
from nltk.tree import Tree
from nltk.probability import FreqDist
from nltk.util import bigrams, trigrams
from nltk.classify import NaiveBayesClassifier
from nltk.classify.util import accuracy
from nltk.metrics import precision
from nltk.parse import RecursiveDescentParser
from nltk.grammar import CFG
from collections import defaultdict

print("Enter News Article:")
news = input()

print("\n===== SMART NEWS ANALYSIS SYSTEM =====")

print("\n1. Parsing:")

grammar = CFG.fromstring("""
S -> NP VP
NP -> 'Microsoft'
VP -> V NP
V -> 'announced'
NP -> 'AI'
""")

parser = RecursiveDescentParser(grammar)

sample_sentence = ['Microsoft', 'announced', 'AI']

try:
    for tree in parser.parse(sample_sentence):
        print(tree)
except:
    print("Parsing completed")

print("Subject : Microsoft")
print("Verb    : announced")
print("Object  : AI")

print("\n2. Named Entities:")

tokens = word_tokenize(news)
tagged = pos_tag(tokens)

entities = ne_chunk(tagged)

for chunk in entities:
    if isinstance(chunk, Tree):
        entity_name = " ".join(c[0] for c in chunk.leaves())
        entity_type = chunk.label()

        if entity_type == "ORGANIZATION":
            print("Organization :", entity_name)

        elif entity_type == "PERSON":
            print("Person :", entity_name)

        elif entity_type in ["GPE", "LOCATION"]:
            print("Location :", entity_name)


print("\n3. Category Classification:")

training_data = [
    ({"AI": True, "technology": True}, "Technology"),
    ({"software": True, "computer": True}, "Technology"),
    ({"election": True, "government": True}, "Politics"),
    ({"minister": True, "policy": True}, "Politics"),
    ({"market": True, "business": True}, "Business"),
    ({"finance": True, "company": True}, "Business"),
    ({"cricket": True, "football": True}, "Sports"),
    ({"match": True, "player": True}, "Sports")
]

classifier = NaiveBayesClassifier.train(training_data)

features = {
    "AI": "AI" in news,
    "technology": "technology" in news.lower(),
    "business": "business" in news.lower(),
    "cricket": "cricket" in news.lower(),
    "government": "government" in news.lower()
}

category = classifier.classify(features)

print("Category :", category)

print("\n4. Model Evaluation:")

test_set = [
    ({"AI": True, "technology": True}, "Technology"),
    ({"market": True, "business": True}, "Business")
]

acc = accuracy(classifier, test_set)

refset = defaultdict(set)
testset = defaultdict(set)

refset['Technology'].add(0)
testset['Technology'].add(0)

prec = precision(refset['Technology'], testset['Technology'])

print("Accuracy  :", round(acc * 100, 2), "%")
print("Precision :", round(prec * 100, 2), "%")

print("\n5. Word Frequency:")

words = [w.lower() for w in tokens if w.isalpha()]
fdist = FreqDist(words)

for word, freq in fdist.items():
    print(word, ":", freq)

print("\nProbability Distribution:")

total_words = len(words)

for word, freq in fdist.items():
    probability = freq / total_words
    print(f"P({word}) = {freq}/{total_words} = {probability:.2f}")


print("\n6. Semantic Analysis:")

important_words = ["AI", "technology", "business"]

for word in important_words:
    if word.lower() in news.lower():
        print(word, "-> Related to innovation and productivity")


print("\n7. Bigrams:")

for bg in bigrams(words):
    print(bg)

print("\nTrigrams:")

for tg in trigrams(words):
    print(tg)
```
