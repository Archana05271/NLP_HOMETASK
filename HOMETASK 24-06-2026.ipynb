```python
from collections import Counter
import nltk
import string
import matplotlib.pyplot as plt
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize


text = input("Enter news articles: ")
text = text.lower()


for p in string.punctuation:
    text = text.replace(p, "")

# 3. Tokenization
words = word_tokenize(text)

# Remove stopwords
stop_words = set(stopwords.words('english'))
filtered_words = [word for word in words if word not in stop_words]

print("\nTokens:")
print(filtered_words)

# 4. Word Frequency Distribution
freq = Counter(filtered_words)

print("\nWord Frequency Distribution:")
for word, count in freq.items():
    print(word, ":", count)

# 5. Vocabulary (Unique Words)
vocabulary = set(filtered_words)

print("\nVocabulary:")
print(vocabulary)

print("\nVocabulary Size:", len(vocabulary))

# 6. Top 5 Most Frequent Words
print("\nTop 5 Frequent Words:")

for word, count in freq.most_common(5):
    print(word, ":", count)

# 7. Type Token Ratio (TTR)
total_tokens = len(filtered_words)

ttr = len(vocabulary) / total_tokens

print("\nType Token Ratio (TTR):", round(ttr, 2))


# 8. Dominant Topic Detection
topics = { "Artificial Intelligence": ["ai", "artificial", "intelligence", "machine", "learning"], "Healthcare": ["healthcare", "medical", "hospital"],"Education": ["education", "learning", "school"],"Technology": ["technology", "innovation", "tools"]}
topic_scores = {}

for topic, keywords in topics.items():
    score = 0
    for keyword in keywords:
        score += freq[keyword]
    topic_scores[topic] = score

dominant_topic = max(topic_scores, key=topic_scores.get)

print("\nTopic Scores:")
for topic, score in topic_scores.items():
    print(topic, ":", score)

print("\nDominant Topic:", dominant_topic)

plt.figure(figsize=(8,5))
plt.bar(freq.keys(), freq.values())

plt.title("Word Frequency Analysis")
plt.xlabel("Words")
plt.ylabel("Frequency")

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
