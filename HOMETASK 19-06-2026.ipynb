```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk import pos_tag
from nltk.probability import FreqDist

# Sample resume text
text = "Dedicated Software Engineer with strong skills in Java, Data Analytics, Artificial Intelligence, and Cloud Computing. Experienced in developing scalable applications and solving real-world problems."

# 1. Tokenization
tokens = word_tokenize(text)

# 2. Lowercase + Remove Punctuation
tokens = [word.lower() for word in tokens if word.isalpha()]

# 3. Remove Stopwords
stop_words = set(stopwords.words('english'))
filtered_tokens = [word for word in tokens if word not in stop_words]

# 4. Stemming
stemmer = PorterStemmer()
stemmed_words = [stemmer.stem(word) for word in filtered_tokens]

# 5. Lemmatization
lemmatizer = WordNetLemmatizer()
lemmatized_words = [lemmatizer.lemmatize(word) for word in filtered_tokens]

# 6. POS Tagging
pos_tags = pos_tag(filtered_tokens)

# 7. Frequency Distribution
freq_dist = FreqDist(filtered_tokens)

# Output
print("Tokens:", tokens)
print("\nFiltered Tokens:", filtered_tokens)
print("\nStemmed Words:", stemmed_words)
print("\nLemmatized Words:", lemmatized_words)
print("\nPOS Tags:", pos_tags)
print("\nFrequency Distribution:")
print(freq_dist.most_common())
```
